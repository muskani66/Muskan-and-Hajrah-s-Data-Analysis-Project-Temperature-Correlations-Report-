```python
#This is for city document
#
#
#
import pandas as pd


#load data into a dataframe object:
df = pd.read_csv('/content/GlobalLandTemperaturesByCity.csv.zip') #Pd is the pandas and this helps import our csv as it acts like numpy
#pd.read means that it is reading the csv file (it is uploaded on the drive as well as file folder)


#this part of the code removes the empty data, so when we are finding the count of the countries it doesn't get interrupted by empty cells
df.dropna(axis = 0, how='all', inplace=True) #cleans up the ROWS where data is empty
df.dropna(axis = 1, how='all', inplace=True) #cleans up the COLUMNS where data is empty


#find the most common country within the country column
#value_counts() counts the number of occcurences of each value in a dataframe column #https://saturncloud.io/blog/python-pandas-convert-valuecounts-output-to-dataframe/#:~:text=value_counts()%20method%20is%20a,count%20is%20the%20corresponding%20value.
lol = df['Country'].value_counts().idxmax()
count = df['Country'].value_counts().max()

#prints the most common response in the country column
print("The most common Country: " + str(lol) + " with " + str(count) + " respondents using it")


#find most to least countries

count = {} #empty variable for the counts of the countries


#this part of the code loops through the the country column
for response in df['Country'].dropna():  # Skip any empty cells
    country_list = response.split(',') #seperates each country in the list with a comma
    #loops thorugh each country in the list
    for l in country_list: #checks if the country is already in the list
        l = l.strip() #removes empty cells
        if l in count:#if the country is already in the list and has a count and shows up again, add one to it
            count[l] += 1
        else: #if a country shows up which is not in the list, set its value to one
            count[l] = 1

#sorts the countries so that it prints the countries from most to least based on their counts
for l in sorted(count, key=count.get, reverse=True):
    print(l, count[l])
```

```python
#Correlation 1: Graph 1
#this graph helps us to find a concise date range to backup our correlations (since we can't use 200-300 years of data)

#Russia - finding date range
import pandas as pd
import matplotlib.pyplot as plt

#load data into a dataframe object:
df_country = pd.read_csv('/content/GlobalLandTemperaturesByCountry.csv.zip')

#this filters the data so that it only extracts data from when the entry in the country column is russia
df_russia_country = df_country[df_country['Country'] == 'Russia']

#extracts the dates from entries that include russia
df_russia_country['dt'] = pd.to_datetime(df_russia_country['dt'], errors='coerce')

#pd.to_numeric makes it so that 'average temperature' column is numeric (all number values), and this code also ensures empty values are not used
#errors = 'coerce' - invalid parsing will be set as NaN (https://www.educative.io/answers/what-is-the-tonumeric-function-in-pandas)
df_russia_country['AverageTemperature'] = pd.to_numeric(df_russia_country['AverageTemperature'], errors='coerce')

#extracts the data so that dates only after 1900 and 1900 are used, dates before are removed
df_russia_country = df_russia_country[df_russia_country['dt'].dt.year >= 1900]

#removes rows where the average temp values are missing (important to ensure it doesn't mess with the data and has inaccurate values)
df_russia_country.dropna(subset=['AverageTemperature'], inplace=True)

#split the dates into ranges of 10 years
#also formats the decade ranges to contain '-'
df_russia_country['DecadeRange'] = (df_russia_country['dt'].dt.year // 10) * 10
df_russia_country['DecadeLabel'] = df_russia_country['DecadeRange'].astype(str) + '-' + (df_russia_country['DecadeRange'] + 9).astype(str)

#calculates the
df_decade_range = df_russia_country.groupby('DecadeLabel')['AverageTemperature'].agg(lambda x: x.max() - x.min()).reset_index()

#this plots the average temperature for each date range
plt.figure(figsize=(12, 6)) #for size of graph
plt.bar(df_decade_range['DecadeLabel'], df_decade_range['AverageTemperature'], color='pink') #creates the bar and details for bar, like the colour of it
plt.xlabel("Decade Range") # labels the x axis as recade range
plt.ylabel("Temperature Range (°C)") #labels the y axis as temperature range
plt.title("10 year intervals of russia affected by average temp ") #title of the graph
plt.xticks(rotation=45) #rotates the date ranges so they aren't clustered
plt.tight_layout() #this is so that nothing overlaps
plt.show() #plots the graph
```

```python
#Correlation 1: Graph 4


#Russia (1970 to 1979)

import pandas as pd
import matplotlib.pyplot as plt

# load data into a dataframe object:
df = pd.read_csv('/content/GlobalLandTemperaturesByCountry.csv.zip')


# this filters the data so that it's just for russia
df_russia_country = df[df['Country'] == 'Russia']

# extracts the dates from entries that include russia
df_russia_country['dt'] = pd.to_datetime(df_russia_country['dt'], errors='coerce')

# extracts the data only from 1970 and 1979
x1 = df_russia_country['dt'] >= '1970-01-01' #the starting date of the extraction
x2 = df_russia_country['dt'] <= '1979-12-31' #the ending date of the extraction
x = x1 & x2 #now to create a final variable, x, we inputted x1, which is the starting date, and x2, which is the ending date, so that we have that decade range for x
df_russia_country = df_russia_country[x]

# Calculate the average temperature for each year - also finds average temperature for each year using ".mean()"
df_russia_avg = df_russia_country.groupby(df_russia_country['dt'].dt.year)['AverageTemperature'].mean()

#plotting the graph
plt.figure(figsize=(10, 6))#size of graph
plt.bar(df_russia_avg.index, abs(df_russia_avg.values), color='pink') #bars of the average temps in the specfic years from 1970-1979, also colour of it
plt.xlabel("Year") #labelled the x axis as year
plt.ylabel("Average Temperature (°C)") #labelled the y axis as average temo
plt.title("Average Temperature in Russia (1970-1979)") #title of graph
plt.xticks(rotation=45) #rotates the values (so years), on the x axis so they don;t look overlapped
plt.show() #plot the graph
```

```python
#Correlation 1: Graph 9


#The line plot Avg tempreture for brazil, india, and russia:


#I essentially just used the same idea but tweaked it up


import plotly.express as px
import pandas as pd
df = pd.read_csv('/content/GlobalLandTemperaturesByCountry.csv')

#getting the year with w3schools https://stackoverflow.com/questions/1133147/how-to-extract-the-year-from-a-python-datetime-object
df['dt'] = pd.to_datetime(df['dt'], errors='coerce')

year = pd.DatetimeIndex(df['dt']).year
df['Year'] = year

#This is just for india, russia, brazil
country_we_want = ['India', 'Russia', 'Brazil']
df = df[df['Country'].isin(country_we_want)]

#This is for the custom data from 1970-1979
df_lol = df[df['dt'].dt.year.between(1970, 1979)]  #Instead of the x1 and x2 function we used this between function which we only found later, its a much more strategic and better way , also because when we kept the normal x1 and x2 we kept getting an error, we used a bit of gemini to help with what was happening wrong but we didn't copy
#https://pandas.pydata.org/docs/reference/api/pandas.Series.between.html #we got more info here on it

#because this is a line plot you can't technically have a scale as it shows for the whole duration so what we ended up doing was that we had to ensure that for every country per year they only have 1 value by getting the mean, the msot average
#What the as_index=false is doing is that it keeps the grouped data that we did below with the mean still seperate colomns however for each colomns it just gets the avergae
#https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.groupby.html
df_loll = df_lol.groupby(['Year', 'Country'], as_index=False)['AverageTemperature'].mean()


#https://plotly.com/python/animations/ - this info from here - took out log
#we took out most of the things that were from the scatter plot because therewas no scale to make and also we added line_group which we saw from the same site if u scroll down
#we use markers for the circle points to show each year
fig = px.line(df_loll, x = "Year", y = "AverageTemperature", color="Country", title="Avg temperture for brazil, india, and russia", line_group="Country", markers=True)

fig.show()
```

```python
#Correlation 2: Graph 1


#The bar plot Avg tempreture in winterfor brazil, india, and russia:


#Same code as scatter but just did a bar and added colour scale



import plotly.express as px
import pandas as pd
df = pd.read_csv('/content/GlobalLandTemperaturesByCountry.csv')

#getting the year with w3schools https://stackoverflow.com/questions/1133147/how-to-extract-the-year-from-a-python-datetime-object
df['dt'] = pd.to_datetime(df['dt'], errors='coerce')

year = pd.DatetimeIndex(df['dt']).year
df['Year'] = year

#This is just for india, russia, brazil
country_we_want = ['India', 'Russia', 'Brazil']
df = df[df['Country'].isin(country_we_want)]

#making the winter interval
df_winter = df[(df['dt'].dt.month == 12) | (df['dt'].dt.month <= 2)]

#This is for the custom data from 1970-1979
x1 = df_winter['dt'] >= '1970-01-01'
x2 = df_winter['dt'] <= '1979-12-31'
x = x1 & x2
df_winter = df_winter[x]

#https://plotly.com/python/builtin-colorscales/ - colour scale - love this site its amazing - we used this for the scale of the colour - its called color_countinuous scale
#https://plotly.com/python/animations/ - this info from here
fig = px.bar(df_winter, x = "Country", y = "AverageTemperature", animation_frame = "Year", animation_group = "Country",
           color="AverageTemperature", color_continuous_scale= "ice", title="Avg Winter temperture for brazil, india, and russia", range_y=[df_winter['AverageTemperature'].min(), df_winter['AverageTemperature'].max()])

fig.show()
```

```python
#Correlation 2: Graph 2


#The line plot Avg tempreture in winterfor brazil, india, and russia:


#same code as the line plot above but I added a blue lining to represent winter


import plotly.express as px
import pandas as pd
df = pd.read_csv('/content/GlobalLandTemperaturesByCountry.csv')

#getting the year with w3schools https://stackoverflow.com/questions/1133147/how-to-extract-the-year-from-a-python-datetime-object
df['dt'] = pd.to_datetime(df['dt'], errors='coerce')

year = pd.DatetimeIndex(df['dt']).year
df['Year'] = year

#This is just for india, russia, brazil
country_we_want = ['India', 'Russia', 'Brazil']
df = df[df['Country'].isin(country_we_want)]

#making the winter interval
df_winter = df[(df['dt'].dt.month == 12) | (df['dt'].dt.month <= 2)] #specified the months of winter
df_winter = df_winter[df_winter['dt'].dt.year.between(1970, 1979)] #https://pandas.pydata.org/docs/reference/api/pandas.Series.between.html
df_winter_avg = df_winter.groupby(['Year', 'Country'], as_index=False)['AverageTemperature'].mean()

#https://plotly.com/python/animations/ - this info from here - took out log
fig = px.line(df_winter_avg, x = "Year", y = "AverageTemperature", color="Country", title="Avg Winter temperture for brazil, india, and russia over the months from december to feburary", line_group="Country", markers=True)

fig.update_traces(line=dict(color="blue")) #https://plotly.com/python/marker-style/ - making it blue for the winter graph

fig.show()
```

```python
#Correlation 2: Graph 3


#The line plot Avg tempreture in summer for brazil, india, and russia:


#same code as Correlation 2: Graph 2 just diff colours and months


#The line plot Avg tempreture in winterfor brazil, india, and russia:

import plotly.express as px
import pandas as pd
df = pd.read_csv('/content/GlobalLandTemperaturesByCountry.csv')

#getting the year with w3schools https://stackoverflow.com/questions/1133147/how-to-extract-the-year-from-a-python-datetime-object
df['dt'] = pd.to_datetime(df['dt'], errors='coerce')

year = pd.DatetimeIndex(df['dt']).year
df['Year'] = year

#This is just for india, russia, brazil
country_we_want = ['India', 'Russia', 'Brazil']
df = df[df['Country'].isin(country_we_want)]

#making the winter interval
df_summer = df[(df['dt'].dt.month >= 6) | (df['dt'].dt.month <= 8)]
df_summer = df_summer[df_summer['dt'].dt.year.between(1970, 1979)] #https://pandas.pydata.org/docs/reference/api/pandas.Series.between.html
df_summer_avg = df_summer.groupby(['Year', 'Country'], as_index=False)['AverageTemperature'].mean()

#https://plotly.com/python/builtin-colorscales/ - colour scale - love this site its amazing
#https://plotly.com/python/animations/ - this info from here - took out log
fig = px.line(df_summer_avg, x = "Year", y = "AverageTemperature", color="Country", title="Avg summer temperture for brazil, india, and russia from months of june until august", line_group="Country", markers=True)

fig.update_traces(line=dict(color="red")) #https://plotly.com/python/marker-style/ - making it blue for the winter graph

fig.show()
```

```python
#Correlation 3: Graph 1



# world map - for third comparison
#searched up on google the percentages of how rural and urban these countries are (added niger and netherlands because
#they are on the higher of being rural or urban) - this'll help to back up our data more accurately
# russia - urban (75%)
# india - rural (64%)
# brazil- urban (87%)
# Niger - rural (83%)
# Netherlands - urban (90%)

#/content/GlobalLandTemperaturesByCountry.csv.zip

# map data from: https://www.naturalearthdata.com/downloads/110m-cultural-vectors/
# graph and code from: https://naturaldisasters.ai/posts/python-geopandas-world-map-tutorial/#download-the-code
#inspo: https://geopandas.org/en/v0.9.0/docs/user_guide/mapping.html


import os
import pandas as pd
import matplotlib.pyplot as plt
import geopandas as gpd
import plotly.express as px


# Upload shape of map
shape_map = "/content/ne_110m_admin_0_countries.shp"

# upload the csv file, assign it to variable
#loading csv file into dataframe
upload_file = '/content/GlobalLandTemperaturesByCountry.csv.zip'
data = pd.read_csv(upload_file)

# from the date column remove any invalid entries (like blank ones)
data['dt'] = pd.to_datetime(data['dt'], errors='coerce')
data['Year'] = data['dt'].dt.year  # Extract year from the dates

# countries we are analyzing the average temp changes of
selected_countries = ['India', 'Russia', 'Brazil', 'Niger', 'Netherlands']

# extracts the data so only the countries we selected are gonna be included in the world map
filtered_data = data[data['Country'].isin(selected_countries)]

# the rows where the average temp has an invalid entry are dropped beccause we don't count those in our data
filtered_data = filtered_data.dropna(subset=['AverageTemperature'])

# calcuyklating the average temp so we can get the average temp per year, which in the end will help us to decide what
#colour it'll be each year based on the scale we have on the side
avg_temp_by_year = filtered_data.groupby(['Year', 'Country'], as_index=False).mean()

# creating map, and scale on side to tell the temperatures colour
fig = px.choropleth( #https://plotly.com/python/choropleth-maps/
    avg_temp_by_year,
    locations="Country", #this makes it so that the columns of the country matches the data of the country on the map
    locationmode="country names", #the locations depend on the ciuntry name
    color="AverageTemperature", #colour depends on the average temperature of the country
    hover_name="Country", #what country name will display when u hover over the country with ur cursor
    animation_frame="Year", #adds the scale at the bottom so u can slide through it and go over the eyars
    color_continuous_scale=px.colors.sequential.Plasma, #bar at the side to represent the temp of each colour
    title="Average Temperature Change Over Time", #map title
    range_color=[avg_temp_by_year['AverageTemperature'].min(), avg_temp_by_year['AverageTemperature'].max()] #the meaning of the different colours (the different temps)
)


fig.show() #prints the map
```
